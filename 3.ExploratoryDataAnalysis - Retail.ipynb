```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Segment',y='Profit',hue='Region',data=df);
plt.xticks(rotation=90)
```

```python
# from the above graph it can be seen that Highest Profit was earned by the HomeOffice segment at the Eastern region.
# And lowest profit can be found at the consumer segment at the central region.
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Segment',y='Discount',hue='Region',data=df);
plt.xticks(rotation=90)
```

```python
#The Consumer segment at the central region gives the highest discount.
#This can be a reason why the consumer region earns less profits compared to the other segments at various other regions.
```

```python
# Hence central region should avoid giving higher discounts in order to increase their profits and minimising loss.
```
